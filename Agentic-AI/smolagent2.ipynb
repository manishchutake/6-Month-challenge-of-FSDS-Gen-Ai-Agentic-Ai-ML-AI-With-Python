```python
import os 
import litellm 
import gradio as gr
from smolagents import CodeAgent, DuckDuckGoSearchTool, LiteLLMModel
import google.generativeai as genai
```

```python
pip install gradio
```

```python
litellm._turn_on_debug()
```

```python
os.environ["GOOGLE-API_KEY"] = "" 
genai.configure(api_key = os.getenv("GOOGLE-API_KEY"))
```

```python
model = LiteLLMModel(
    model_id="gemini/gemini-1.5-flash",
    api_key=os.getenv("GOOGLE_API_KEY")
)
```

```python
ml_code_agent = CodeAgent(
    
    tools = [DuckDuckGoSearchTool()],
    additional_authorized_imports = ['pandas', 'numpy','sklearn'],
    model = model 
    )
```

```python
def generate_ml_code(prompt):
    full_prompt = """"
    create a step- by- step plan to write python code for svm model using sklearn to classify the iris dataset 
"""
    try:
        response = ml_code_agent.run(full_prompt)
        return response
    except Exception as e:
        return f"Error:{str(e)}"
```

```python
iface = gr.Interface(
    fn = generate_ml_code,
    inputs = "text",
    outputs = "text",
    title = "ML code Generator (Gemini 2.0 Flash)",
    description = "Enter a Machine Learning coding task, and the AI will genenrate python code using"
)
```

```python
iface.launch()
```
